```python
import pandas as pd

#read the data and create a dataframe
neighbourhood = pd.read_csv('listings.csv')

#questionnaire1
#count the listings per neighbourhood
counts = neighbourhood['neighbourhood'].value_counts()

#sort the counts variable in decreasing order
counts = counts.sort_values(ascending = False)

#display the outcome
print(counts)
```

```python
#questionnaire2
#group the dataframe obj by neighbourhood
geitonies = neighbourhood.groupby('neighbourhood')

#use panda functions to calculate the statistics
times = geitonies['price'].agg(['median','mean','std','count'])

#sort them by their median indescending order
sorted_times = times.sort_values(by = 'median', ascending = False)

#print the data
print(sorted_times)
```

```python
#questionnaire3
#we will need the matplotlib library for this questionnaire
#numpy goes hand in hand with matplotlib
import matplotlib.pyplot as plot
import numpy as np

#read the calendar.csv file that contains the availability per day for each listing
calendar_file = pd.read_csv('calendar.csv')

#convert the date column into a datetime object
calendar_file['date'] = pd.to_datetime(calendar_file['date'])

#filter out the unavailable dates
availables = calendar_file[calendar_file.available == 't']

#group the listings per date and count per date
counts = availables.groupby('date')['listing_id'].count()

#now we have to create a plot object to display the data
plot.figure(figsize=(20,10))
plot.bar(counts.index, counts.values)
plot.title('Available listings per Date')
plot.xlabel('Date')
plot.ylabel('Number of Listings')
plot.xticks(rotation=45)
plot.tight_layout()

#show the plot
plot.show()
```

```python
#questionnaire 4 part 1
#read the data and create a dataframe
listings = pd.read_csv('listings.csv')

#find the lowest and higher number of reviews within the listings file
min = listings.number_of_reviews.min()
max = listings.number_of_reviews.max()

# Calculate the maximum number of reviews rounded up to the closest hundred
bin = np.arange(min // 100 * 100, (max // 100 + 2) * 100, 100)

# Create a histogram
plot.figure(figsize=(20, 10))
plot.hist(listings['number_of_reviews'], bins = bin, edgecolor='k')
plot.title('Number of Reviews per Listing')
plot.xlabel('Listings')
plot.ylabel('Reviews')
plot.xticks(np.arange(min, max + 101, 100), rotation=45)
plot.grid(axis='y', alpha=0.75)

# Display the plot
plot.show()
```

```python
#part 2 of questionnaire4
#calculate the average reveiews per month
average_reviews = listings.reviews_per_month.mean()

#calculate average occupancy
occupancy = average_reviews * 2
print(occupancy)
```

```python
#part3 of questionnaire4
#calculate the average price for all listings
average_price = listings.price.mean()

#calculate the average income per listing for 30 days for 3 people per night
average_income = average_price *3 * 30
print(average_income)
```

```python
#questionnaire 5
#showing the number of room types
room_types = listings.room_type.value_counts()
print(room_types)
```

```python
#questionnaire 5
#showing the number of room types in percentages
room_types_percentages = listings.room_type.value_counts(normalize = True)* 100
print(room_types_percentages)
```

```python
#questionnaire 5
#count the listings per neighbourhood and room type at the same time
var = listings.groupby(['neighbourhood','room_type']).size()
print(var)
```

```python
#questionnaire 6
#open the file
calendar = pd.read_csv('calendar.csv')

# Convert the 'date' column to a datetime object
calendar['date'] = pd.to_datetime(calendar['date'])

#we have to convert the string prices into float prices
calendar['price'] = pd.to_numeric(calendar['price'].str.replace('[^\d.]', '', regex=True), errors='coerce')

# Group the data by 'date' and calculate the average price per listing per day
grouped_days = calendar.groupby('date')
average = grouped_days['price'].agg('mean')
print(average)

#now create a graph to visualize the data
plot.figure(figsize=(12, 6))
plot.plot(average.index, average.values, marker='o', linestyle='-')
plot.title('Average Price per Listing per Day')
plot.xlabel('Date')
plot.ylabel('Average Price')
plot.xticks(rotation=45)
plot.grid()

# Display the graph
plot.show()
```

```python
#last questionnaire
#first we have to import the folium library
import folium
from folium.plugins import MarkerCluster
import pandas as pd

#read the listings csv again
listings = pd.read_csv('listings.csv')

#set athens center location
map = folium.Map(location=[37.9838, 23.7275], zoom_start=12)  # You can adjust the zoom level

#add the same loc listings in a cluster, so that your pc doesnt explode when running the html file
cluster = MarkerCluster().add_to(map)

#iterate through the listings and add them to the cluster instead of the map, so that our pc doesnt explode
for index, row in listings.iterrows():  # Use index and row to iterate through the DataFrame
    popup_content = f"<strong>ID:</strong> {row['id']}<br><strong>Name:</strong> {row['name']}<br><strong>Price:</strong> {row['price']}"
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup = popup_content
    ).add_to(cluster)
    
#save the map in an external html file
map.save("map2.html")
```
